### working with csv files

```python
# importing pandas
import pandas as pd
```

```python
# opening a local csv
df = pd.read_csv('aug_train.csv')
df.head()
```

```python
# Opening a csv file from an URL 
# or server se kese csv file ka data load karne ke liye
import requests
from io import StringIO

url = "https://raw.githubusercontent.com/campusx-official/100-days-of-machine-learning/refs/heads/main/day15%20-%20working%20with%20csv%20files/aug_train.csv"
headers ={"User-Agent":"Mozilla/5.0 (Macintosh;Intel Mac OS X 10.14; rv:66.0) Gecko/20100101 Firefox/66.0"}
req = requests.get(url,headers=headers)
data = StringIO(req.text)

df = pd.read_csv(data)

df.head()
```

### Sep Parameter

```python
# rather then finding comma find tab and uske base par column create karo
pd.read_csv('movie_titles_metadata.tsv',sep='\t',names=['sno','name','release-year','rating','votes','genres'])
```

### Index_col parameter

```python
pd.read_csv('aug_train.csv',index_col='enrollee_id')
```

### Header parameter

```python
pd.read_csv('test.csv',header=1)
```

### use_cols parameter


```python
pd.read_csv('aug_train.csv',usecols=['enrollee_id','city'])
```

### squeeze parameters

```python
df =pd.read_csv('aug_train.csv',usecols=['enrollee_id']).squeeze()
type(df)
df
```

### Skiprows/nrows Parameter

```python
pd.read_csv('aug_train.csv',skiprows=[0,1])
pd.read_csv('aug_train.csv',nrows=100)
```

### Encoding parameter

```python
pd.read_csv('zomato.csv',encoding='latin-1')
```

### Skip bad lines

```python
pd.read_csv('zomato.csv', encoding='latin-1', on_bad_lines='skip')
```

### dtype parameter

```python
pd.read_csv('aug_train.csv',dtype={'target':int}).info()
```

### Handling Dates

```python
# by default date ko as a string pass kiya jata hai but apko usko date time format mai rakhna padhta hai
pd.read_csv('ipl-matches.csv',parse_dates=['Date']).info()
```

### Convertors

```python
def rename(name):
    if name == "Rajasthan Royals":
        return "RR"
    else:
        return name
```

```python
rename('Rajasthan Royals')
```

```python
pd.read_csv('ipl-matches.csv',converters={'Team1':rename})
```

### na_values parameter

```python
pd.read_csv('aug_train.csv',na_values=['Male'])
```
